# Duplicate Analysis of AcousticBrainz Submissions

There are many duplicate submissions in AcousticBrainz, files that represent the same music recording (and have the same MBID) but were submitted by many people. These files could be from different sources (e.g. different CDs issues, a remastered CD, a recorded vinyl, a radio edit) and could be encoded using different formats (MP3, Flac, AAC). A dataset of only songs with more than 1 submission has been made available [here](http://www.dtic.upf.edu/~aporter/amplab/).

Also, there are some submissions in AcousticBrainz that have been incorrectly labeled. That is, the contents of the file which was analysed is not the file which the MBID label indicates. The goal of this task is to find which submissions in the provided dataset are mislabeled. In this context, it will be helpful to know which of the submissions were duplicates for sure. Hence, I would like to propose an analysis approach which will tell us most certain duplicates and most certain non-duplicates.

The descriptors - 'length', 'bpm', 'average loudness', 'onset rate', 'key_key', 'key_scale', 'replaygain' and 'tuning frequency' have been considered. We then rejected some of these descriptors by estimating how useful they are.

The dataset for this task has been made available as an archive of json files, in the format that AcousticBrainz uses to store the data. Since, I did not need all of the data included in these files, I reduced the data to just the necessary values, and then worked with that data. I then looped through each of the files, extracted each of these features, and stored them in a CSV file.

```python
#This class can be used to instantiate an object which accepts a dictionary
#and allows us to access the contents of nested json through its 'get' function.
#Source: https://www.haykranen.nl/2016/02/13/handling-complex-nested-dicts-in-python/
class DictQuery(dict):
    def get(self, path, default = None):
        keys = path.split("/")
        val = None

        for key in keys:
            if val:
                if isinstance(val, list):
                    val = [ v.get(key, default) if v else None for v in val]
                else:
                    val = val.get(key, default)
            else:
                val = dict.get(self, key, default)

            if not val:
                break;

        return val
```

## Defining 'Ground Truth'

This step is based on our assumption that if multiple entries for a MBID have the same values for "critical' descriptors, then they are duplicates and these descriptor values can be used as the ground truth for that particular MBID. We then specify tolerence for deviation from these critical descriptor feature vector to decide whether the entry is duplicate or not.

### Discussion:
#The shortlisted_features need to be such that we have maximum duplicity of the combination of these values across all the MB_IDs. There are 1026 unique MB_IDs. The idea is to select the shortlisted_features set such that the frequency distribution of groundtruth has lowest kurtosis("peakiness") : "ground_truth\['count'\].kurtosis()" 

However, we have a constraint.The confidence of our analysis will be depending on the number of features that we select in the shortlisted_features set. With more features, we notice that the kurtosis increases. In this way, we need to find the optimum trade-off between the number of features and kurtosis of the frequency distribution of the presumed duplicates.


```python
shortlisted_features=['id','bpm','lenght','key_key']

# Get frequency distribution grouped by shortlisted_features
frequency_distribution=dfv.groupby(shortlisted_features).size().reset_index().rename(columns={0:'count'}).sort_values('count',ascending=False)
# This has entries with repeated MBIDs. We need to retain only the ones with highest counts.
total_mb_id=len(frequency_distribution['id'].unique())
mb_id=set()
counter=0
shortlisted_features.append('count')
groundTruthValues={}
ground_truth=pd.DataFrame(groundTruthValues, columns=shortlisted_features)
for index in range(len(frequency_distribution["count"])):
    if counter==total_mb_id:
        break
    row = frequency_distribution.iloc[index,:]
    #We break once we have covered all the unique MB_IDs
    if row['id'] not in mb_id:
        mb_id.add(row['id'])
        ground_truth.loc[counter]=row
        counter+=1
        
#Kurtosis(peakiness) can be computed using this:
#ground_truth['count'].kurtosis()
```

## Evaluation

We configure the tolerance parameters for deviation in our selected_features set. The parameters bpm and length are numerical and hence have numerical tolerance bands. However, key is non-numerical. We consider any change is key with repect to the ground truth as a deviation beyond tolerance. The deviation is computed with respect to the ground truth that we established in the previous cell for a MB_ID and is very specific to the MB_ID. The entries which are beyond-the-tolerance-limit in all the parameters of our selected_features set are very highly likely to be non-duplicates than others. For all other conditions like beyond-the-tolerance deviation in one of the paramters or two of them, we can not comment whether they are duplicate or not. However, we know that two submissions with same MB_IDs are more likely to be non-duplicates when they have two beyond-the-tolerance-limit parameters than one. Using this logic we compute the probabilities of entry being non-duplicate.

For example, consider three "shortlisted_features" parameters from ground truth {G1, G2, G3} for an entry E. 
E={G1, G2, G3} 
X={X1, X2, X3} 
Let T={T1, T2, T3} be the tolerance threshold for the features.
Let S be a set of 0 or 1 values {S1, S2, S3} such that Si=(abs(Gi-Xi))>Ti 

Priority values(positive) for each feature: P={P1, P2, P3} such that P1+P2+P3=1  
We define a score function Score(P,S)=sum(Pi\*Si)
Higher the score, higher the probability that the entry is not a duplicate.
The maximum score can be 1 and minimum 0.

```python
#Set the thresholds for tolerance
length_tolerance=30 # in seconds
bpm_tolerance=5 # in beats per minute

trimmed_shortlisted_features=shortlisted_features[1:-1]#Discard the "id" and "count" fields
priority = [0.33,0.34,0.33]#Assign priority to each of the features such that their sum is 1
nonDuplicityProbability = []
print("Time to sit back and relax for 2 minutes!")
# Next, we evaluate the score for every submission in our dataset .. 
# .. by comparing the features with it's corresponding ground truth
for index in range(len(dfv['id'])):
    row = dfv.iloc[index,:]
    mb_id = row['id']
    ground_truth_values=ground_truth[ground_truth['id'] == mb_id]
    gt_bpm=ground_truth_values['bpm'].iloc[0]
    gt_lenght=ground_truth_values['lenght'].iloc[0]
    gt_key=ground_truth_values['key_key'].iloc[0]
    nonDuplicityProbability.append((
        ( abs(gt_bpm - row['bpm'])> bpm_tolerance)*priority[0]
        +(abs(gt_lenght - row['lenght'])> length_tolerance)*priority[1]
        +(gt_key != row['key_key'])*priority[2]))
print("Evaluation completed successfully!")
```

```python
dfv['nonDuplicityProbability']=nonDuplicityProbability
frequencies=list(dfv['nonDuplicityProbability'].value_counts().sort_index())

#['duplicates','inconclusive','non_duplicates']
duplicates=frequencies[0]
inconclusive=sum(frequencies[1:-1])
non_duplicates=frequencies[-1]
values=[duplicates,inconclusive,non_duplicates]
labels=['duplicates','inconclusive','non_duplicates']
plt.figure(figsize=(15,7))

# create a list to collect the plt.patches data
totals = []
ax=sns.barplot(labels, values, alpha=0.8)
ax.tick_params(axis='y', labelsize=25)
ax.tick_params(axis='x', labelsize=30)
for i in ax.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+0.3, i.get_height()+1.5, \
            str(round((i.get_height()/total)*100, 2))+'%', fontsize=25,
                color='dimgrey')
plt.title('NonDuplicity Distribution', fontsize=25)
plt.ylabel('Number of Occurrences', fontsize=25)
plt.xlabel('NonDuplicity', fontsize=25)
plt.savefig(path_to_plots+'NonDuplicity.png')
plt.show()

#unique value count distribution
unique_values=frequencies
unique_labels=sorted(list(dfv['nonDuplicityProbability'].unique()))
plt.figure(figsize=(15,7))

# create a list to collect the plt.patches data
totals = []
ax=sns.barplot(unique_labels, unique_values, alpha=0.8)
ax.tick_params(axis='y', labelsize=25)
ax.tick_params(axis='x', labelsize=30)
for i in ax.patches:
    totals.append(i.get_height())

# set individual bar lables using above list
total = sum(totals)

# set individual bar lables using above list
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+0.2, i.get_height()+1.5, \
            str(round((i.get_height()/total)*100, 2))+'%', fontsize=25,
                color='dimgrey')
plt.title('Unique Probability Values Distribution', fontsize=25)
plt.ylabel('Number of Occurrences', fontsize=25)
plt.xlabel('Unique Probability Values', fontsize=25)
plt.savefig(path_to_plots+'UniqueProbabilityValues.png')
plt.show()

```
