### Homework 4 : Evaluation of various onset detection methods from Essentia

This notebook contains all steps of downloading data, extracting, running tests and reports. We make use of the onset detection functions: 'HFC', 'complex', 'complex_phase' and 'fluxâ€™ from Essentia and then use [mir_eval](https://github.com/craffel/mir_eval) library to evaluate its performance.


```python
#All the imports are done here
import os,sys
import urllib.request
import zipfile
from essentia import *
from essentia.standard import *
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import get_window
import mir_eval
from essentia import Pool, array
import IPython.display as ipd
```

```python
#Download data from URL
url='http://www.rhythmos.org/shareddata/OnsetLabeledInstr2013.zip'
filename='OnsetLabeledInstr2013.zip'
targetDir='../data/onsets/'
os.umask(0) #To mask the permission restrictions on new files/directories being created
os.makedirs(targetDir)
finalDataDir=targetDir+filename.replace('.zip','')

#Downloading the zip file from the url
urllib.request.urlretrieve(url,filename)
#Unzipping to a specific folder
zip_ref = zipfile.ZipFile(filename, 'r')
zip_ref.extractall(targetDir)
zip_ref.close()
os.remove(filename)#Removing the zip file
print('Data downloaded and unzipped to: ',finalDataDir)
```

```python
files=os.listdir("../data/onsets/Percussion")
soundsDir='../data/onsets/Percussion'

wavFileNames=[]
for file in files:
    if file.endswith('.wav'):
        wavFileNames.append(file)
fs = 44100
durInSecs=2
t=np.arange(durInSecs*fs)/float(fs)
zero_array=t*0#used only for plotting purposes
windowSize=1024
hopSize=256

#Loading Essentia functions required
# See Essenatia tutorial example: 
# https://github.com/MTG/essentia/blob/master/src/examples/tutorial/example_onsetdetection.py
od_hfc = OnsetDetection(method='hfc')
od_complex = OnsetDetection(method='complex')
od_complex_phase = OnsetDetection(method='complex_phase')
od_flux = OnsetDetection(method='flux')
w = Windowing(type = 'hann')
fft = FFT() # this gives us a complex FFT
c2p = CartesianToPolar() # and this turns it into a pair (magnitude, phase)
onsets = Onsets()
```

```python
ind=0
for file in wavFileNames:
    fileName = os.path.join(soundsDir, file)
    x = MonoLoader(filename = fileName, sampleRate = fs)()
    #x = x[:durInSecs*fs]#let's use only the first two seconds of the signals
    x = x/np.max(np.abs(x))
    #reading ground-truth annotation
    true_onsets=np.loadtxt(fileName.replace('.wav','.txt'))
    
    t=np.arange(len(x))/float(fs)
    zero_array=t*0#used only for plotting purposes
 
    #Plotting
    f, axarr = plt.subplots(5,1,figsize=(13, 5))
    axarr[0].plot(t,x);axarr[0].set_title(file);axarr[0].axis('off')
    axarr[0].vlines(true_onsets, -1, 1, color='r')
    
    #Essentia beat tracking
    pool = Pool()
    for frame in FrameGenerator(x, frameSize = 1024, hopSize = 512):
        mag, phase, = c2p(fft(w(frame)))
        pool.add('features.hfc', od_hfc(mag, phase))
        pool.add('features.complex', od_complex(mag, phase))
        pool.add('features.complex_phase', od_complex_phase(mag, phase))
        pool.add('features.flux', od_flux(mag, phase))
    
    onsets_hfc = onsets(array([pool['features.hfc']]),[1])
    #onsets_hfc = onsets_hfc[onsets_hfc<durInSecs]
    np.savetxt(fileName.replace('.wav','_est_hfc.txt'), onsets_hfc, delimiter='\n',fmt='%f')
    
    onsets_complex = onsets(array([pool['features.complex']]), [1])
    #onsets_complex = onsets_complex[onsets_complex<durInSecs]
    np.savetxt(fileName.replace('.wav','_est_complex.txt'), onsets_complex, delimiter='\n',fmt='%f')
    
    onsets_complex_phase = onsets(array([pool['features.complex_phase']]), [1])
    #onsets_complex_phase = onsets_complex[onsets_complex<durInSecs]
    np.savetxt(fileName.replace('.wav','_est_comphase.txt'), onsets_complex_phase, delimiter='\n',fmt='%f')
    
    onsets_flux = onsets(array([pool['features.flux']]), [1])
    #onsets_complex = onsets_complex[onsets_complex<durInSecs]
    np.savetxt(fileName.replace('.wav','_est_flux.txt'), onsets_flux, delimiter='\n',fmt='%f')
    
    #Plotting onsets computed by essentia 
    axarr[1].plot(t,zero_array);axarr[1].set_title('Onsets detected using HFC');
    axarr[1].axis('off')
    axarr[1].vlines(onsets_hfc, -1, 1, color='b')
    
    axarr[2].plot(t,zero_array);axarr[2].set_title('Onsets detected using complex spectral difference');
    axarr[2].axis('off')
    axarr[2].vlines(onsets_complex, -1, 1, color='b')
    
    axarr[3].plot(t,zero_array);axarr[3].set_title('Onsets detected using complex phase spectral difference');
    axarr[3].axis('off')
    axarr[3].vlines(onsets_complex_phase, -1, 1, color='b')
    
    axarr[4].plot(t,zero_array);axarr[4].set_title('Onsets detected using difference in spectral magnitudes');
    axarr[4].axis('off')
    axarr[4].vlines(onsets_flux, -1, 1, color='b')
    
```

```python
for file in wavFileNames:
    fileName = os.path.join(soundsDir, file)
    #Calculate f_measure averaged across all the tracks 
    reference_beats = mir_eval.io.load_events(fileName.replace('.wav','.txt'))
    estimated_beats = mir_eval.io.load_events(fileName.replace('.wav','_est_hfc.txt'))
    F, P, R = mir_eval.onset.f_measure(reference_beats, estimated_beats)
    print(fileName)
    print("F : "+str(F))
```

# Equal loudness cases give better results. The performance stats displayed above are the scores averaged out across all the 13 samples. When loudness is equalized, we achieve a higher voicing recall, lower false alarm, higher raw pitch accuracy, higher raw chroma accuracy and higher overall accuracy in comparison to the case with no loudness equalized.
